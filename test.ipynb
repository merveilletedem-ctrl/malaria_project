```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

# pour nettoyer les colones invisibles  
df.columns = df.columns.str.strip()

#nettoyer la colone cible
df["No. of cases"] = df["No. of cases"].astype(str).str.extract(r'(\d+)')

# convertir y en nombre 
y = pd.to_numeric(y, errors='coerce')

# supprimer les lignes où la cible est vide
df = df.dropna(subset=["No. of cases"])

# Séparer X et y
X = df.drop("No. of cases", axis=1)
y = df["No. of cases"]

# Transformer le texte en nombres
X = pd.get_dummies(X)

# Remplacer les valeurs manquantes
X = X.fillna(X.mean())

# pour verifier qu'il ne s'agit que des nombres 
print(y.isnull().sum())

# Modèle
model = RandomForestRegressor()
model.fit(X, y)
```

```python
predictions = model.predict(X)
print(predictions[:10])
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y, predictions))
```
